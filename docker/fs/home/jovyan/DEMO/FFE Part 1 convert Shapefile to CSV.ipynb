## FFE Part 1 convert shapefile to CSV 


This is a simple preparation step using just python and goepandas. No Spark here.

```python
import datetime
import geopandas as gpd
import os
```

```python
path = "/geodata"
```

```python
%%time

gdf_polygon = gpd.read_file(os.path.join(path, "buildings_raw.shp"))

gdf = gdf_polygon.copy()

gdf["area"] = gdf['geometry'].area
gdf['centroid'] = gdf['geometry'].centroid
gdf['X'] = gdf.centroid.x
gdf['Y'] = gdf.centroid.y
gdf['d_short'] = gdf_polygon.exterior.distance(gdf)
gdf['d_long'] = gdf['area'] / gdf['d_short']

gdf.to_csv("/home/jovyan/DEMO/buildings_raw.csv")
```
